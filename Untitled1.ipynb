```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
import yfinance as yf

%matplotlib inline
```

```python
# Fetch historical data for Bitcoin and Ethereum
btc = yf.download('BTC-USD', start='2015-01-01', end='2025-01-01')
eth = yf.download('ETH-USD', start='2015-01-01', end='2025-01-01')
smp = yf.download('SPY-USD', start='2015-01-01', end='2025-01-01')

# Display the first few rows of each dataset to inspect the data
print("Bitcoin data:")
print(btc.head())
print("Ethereum data:")
print(eth.head())
print("SMP data:")
print(smp.head())
```

```python
btc.dropna(inplace=True)
btc_data = btc.loc['2019-12-31':, 'Close']
eth_data = eth.loc['2019-12-31':, 'Close']

display(btc_data.head())
display(btc_data.tail())
display(eth_data.head())
display(eth_data.tail())
```

```python
df_crypto = pd.concat([btc_data, eth_data], axis=1).dropna()
df_crypto.columns = ['BTC', 'ETH']
display(df_crypto.head())
display(df_crypto.tail())
```

```python
#localize date index to remove +00:00
df_crypto.index = df_crypto.index.tz_localize(None)
df_crypto.head()
```

```python

# Calculate the percentage change for BTC and ETH
df_crypto['BTC_pct_change'] = df_crypto['BTC'].pct_change()
df_crypto['ETH_pct_change'] = df_crypto['ETH'].pct_change()

display(df_crypto.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_crypto['BTC'], label='BTC Percentage Change')
plt.plot(df_crypto['ETH'], label='ETH Percentage Change')
plt.title('BTC and ETH Percentage Change')
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.legend()
plt.grid(True)
plt.show()
```
